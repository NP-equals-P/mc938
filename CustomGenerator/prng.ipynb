# Testes sem threads

```python
def cpu_read(times=5):
    print("----- Executing CPU reads -----")
    cpu_sum = 0
    for i in range(times):
        cpu = psutil.cpu_percent(interval=1)
        print(f"CPU read {i}: {cpu}")
        cpu_sum += cpu

    print("CPU average: ", cpu_sum/times)
```

```python
psutil.virtual_memory()
```

```python
def mem_read(times=5):
    mem_sum = 0
    print("----- Executing Memory Reads -----")
    for i in range(times):
        mem = psutil.virtual_memory().percent
        print(f"Memory read {i}: {mem}")
        mem_sum += mem
        time.sleep(1)

    print("Memory average: ", mem_sum/times)
```

```python
def disk_read(times=5):
    disk_sum = 0
    print("----- Executing Disk Reads -----")
    for i in range(times):
        disk = psutil.disk_usage('/').percent
        print(f"Disk read {i}: {disk}")
        disk_sum += disk
        time.sleep(1)

    print("Disk average: ", disk_sum/times)
```

```python
def net_read(times=5):
    sent_sum = 0
    rec_sum = 0
    print("----- Executing Network Reads -----")
    for i in range(times):
        net = psutil.net_io_counters()
        print(f"Read {i} => Bytes Sent: {net.bytes_sent}, Bytes Received: {net.bytes_recv}")
        sent_sum += net.bytes_sent
        rec_sum += net.bytes_recv
        time.sleep(1)

    print(f"Average => Sent: {sent_sum/times}, Received: {rec_sum/times}")
```

# Sorteador de bytes de uma notícia

```python
def getBinStr(n):
    return bin(n)[2:]
```

```python
getBinStr(97)
```

```python
def tabBinByte(bin):
    # bin é uma string
    return "0"*(8-len(bin)) + bin
```

```python
def getBinASCII(c):
    binStr = getBinStr(ord(c))
    return tabBinByte(binStr)
```

```python
def getChainFromTxt(text, idx=0, num=1):
    substring = text[idx:idx+num]
    return "".join([getBinASCII(ch) for ch in substring])
```

```python
getChainFromTxt("Bom dia", 2, 3)
```

```python
getChainFromTxt(" ")
```

```python
def getTextFromG1(url):
    driver = Firefox()
    driver.get(url)
    textos = driver.find_elements(By.CLASS_NAME, "content-text__container")
    textos_trim = "".join([t.text for t in textos if len(t.text) > 0])
    textos_trim = "".join(textos_trim.split(" "))
    driver.quit()
    return textos_trim
```

```python
trimT2 = getTextFromG1(url2)
trimT2
```

```python
def reverseString(s):
    idxMax = len(s) - 1
    revStr = ""
    for i in range(idxMax, -1, -1):
        revStr += s[i]
    return revStr
```

```python
reverseString("Bom")
```

```python
def invertBits(n):
    # n está como string e ocupa um byte
    operand2 = '1'*len(n)
    n_inverted = int(n, 2) ^ int(operand2, 2)
    return bin(n_inverted)[2:]
```

```python
invertBits('010010')
```

```python
def getByteChain(text, numBytes):
    remaining_bytes = numBytes
    byteChain = ""
    idx = 0
    max_idx = len(text) - 1
    while remaining_bytes > 0:
        net_read = psutil.net_io_counters()
        bytesOut, bytesIn = net_read.bytes_sent, net_read.bytes_recv
        idx += (-1)**(bytesOut%10) * (bytesIn%1000)
        idx = idx % max_idx # Ou len(text) ?
        curr_chain = getChainFromTxt(text, idx)
        # Último dígito
        if bytesOut%2 != 0:
            curr_chain = '1' + curr_chain[1:]
        # Penúltimo dígito
        if ((bytesOut%100)//10)%2 != 0:
            curr_chain = reverseString(curr_chain)
        # Antepenúltimo dígito
        if ((bytesOut%1000)//100)%2 != 0:
            curr_chain = invertBits(curr_chain)
        byteChain += tabBinByte(curr_chain)
        remaining_bytes -= 1
    return byteChain
        
```

# Sorteador de página de notícia

```python
main_url = "https://g1.globo.com/"
```

```python
def pickNewsPage():
    net_read = psutil.net_io_counters()
    bytesOut, bytesIn = net_read.bytes_sent, net_read.bytes_recv
    driver = Firefox()
    driver.get(main_url)
    links = driver.find_elements(By.CLASS_NAME, "feed-post-link")
    link_idx = (bytesOut%1000)%len(links)
    chosen_link = links[link_idx]
    href = chosen_link.get_attribute('href')
    driver.quit()
    return href
```

```python
pickNewsPage()
```

```python
def g1RandomNumberGenerator(numBytes):
    url = pickNewsPage()
    text = getTextFromG1(url)
    byteChain = getByteChain(text, numBytes)
    return byteChain
```

```python
def separateBytes(num):
    i = 0
    while i < len(num):
        print(num[i:i+8], end=" ")
        i += 8
    print("")
```

```python
randomNumber = g1RandomNumberGenerator(64)
separateBytes(randomNumber)
```

```python
def countFirstOnes(number):
    count = 0
    for i in range(len(number)//8):
        if number[i*8] == '1':
            count += 1
    return count
```
