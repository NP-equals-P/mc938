# Testes sem threads

```python
def cpu_read(times=5):
    print("----- Executing CPU reads -----")
    cpu_sum = 0
    for i in range(times):
        cpu = psutil.cpu_percent(interval=1)
        print(f"CPU read {i}: {cpu}")
        cpu_sum += cpu

    print("CPU average: ", cpu_sum/times)
```

```python
psutil.virtual_memory()
```

```python
def mem_read(times=5):
    mem_sum = 0
    print("----- Executing Memory Reads -----")
    for i in range(times):
        mem = psutil.virtual_memory().percent
        print(f"Memory read {i}: {mem}")
        mem_sum += mem
        time.sleep(1)

    print("Memory average: ", mem_sum/times)
```

```python
def disk_read(times=5):
    disk_sum = 0
    print("----- Executing Disk Reads -----")
    for i in range(times):
        disk = psutil.disk_usage('/').percent
        print(f"Disk read {i}: {disk}")
        disk_sum += disk
        time.sleep(1)

    print("Disk average: ", disk_sum/times)
```

```python
def net_read(times=5, wait=True):
    sent_sum = 0
    rec_sum = 0
    print("----- Executing Network Reads -----")
    for i in range(times):
        net = psutil.net_io_counters()
        print(f"Read {i} => Bytes Sent: {net.bytes_sent}, Bytes Received: {net.bytes_recv}")
        sent_sum += net.bytes_sent
        rec_sum += net.bytes_recv
        if wait:
            time.sleep(1)

    print(f"Average => Sent: {sent_sum/times}, Received: {rec_sum/times}")
```

# Sorteador de bytes de uma notícia

```python
def getBinStr(n):
    return bin(n)[2:]
```

```python
getBinStr(97)
```

```python
def tabBinByte(bin):
    # bin é uma string
    return "0"*(8-len(bin)) + bin
```

```python
def getBinASCII(c):
    binStr = getBinStr(ord(c))
    return tabBinByte(binStr)
```

```python
def getChainFromTxt(text, idx=0, num=1):
    substring = text[idx:idx+num]
    return "".join([getBinASCII(ch) for ch in substring])
```

```python
getChainFromTxt("Bom dia", 2, 3)
```

```python
getChainFromTxt(" ")
```

```python
def getTextFromG1(url):
    driver = Firefox()
    driver.get(url)
    textos = driver.find_elements(By.CLASS_NAME, "content-text__container")
    textos_trim = "".join([t.text for t in textos if len(t.text) > 0])
    textos_trim = "".join(textos_trim.split(" "))
    driver.quit()
    return textos_trim
```

```python
def reverseString(s):
    idxMax = len(s) - 1
    revStr = ""
    for i in range(idxMax, -1, -1):
        revStr += s[i]
    return revStr
```

```python
reverseString("Bom")
```

```python
def invertBits(n):
    # n está como string e ocupa um byte
    operand2 = '1'*len(n)
    n_inverted = int(n, 2) ^ int(operand2, 2)
    return bin(n_inverted)[2:]
```

```python
invertBits('010010')
```

```python
def swapBits(b, idx_1, idx_2):
    idx_1, idx_2 = idx_1%len(b), idx_2%len(b)
    if idx_1 > idx_2:
        idx_1, idx_2 = idx_2, idx_1
    bit1, bit2 = b[idx_1], b[idx_2]
    if idx_1 != idx_2:
        return b[:idx_1] + bit2 + b[idx_1+1:idx_2] + bit1 + b[idx_2+1:]
    return b
```

```python
def selectDigit(n, pos : int):
    n_str = str(n)
    digit = n_str[pos%len(n_str)]
    return int(digit)
```

```python
def updateDigit(n, idx, idx_aux):
    digit = selectDigit(n, idx)
    aux = selectDigit(n, idx_aux)
    return (idx-digit)%len(str(n)) - aux + ((aux%2)==0), digit
```

```python
def getByteChain(text, numBytes, debugIdx=False):
    remaining_bytes = numBytes
    byteChain = ""
    idx = 0
    max_idx = len(text) - 1
    i_1, i_2, i_3 = -1, -2, -4
    while remaining_bytes > 0:
        net_read = psutil.net_io_counters()
        bytesOut, bytesIn = net_read.bytes_sent, net_read.bytes_recv
        for i in [i_1, i_2, i_3]:
            bytesOut = swapBits(str(bytesOut), i, -i)
        bytesOut = int(bytesOut)
        i_1, d_1 = updateDigit(bytesOut, i_1, i_3-i_2)
        i_2, d_2 = updateDigit(bytesOut, i_2, i_1-i_3)
        i_3, d_3 = updateDigit(bytesOut, i_3, i_2-i_1)
        idx += (-1)**(d_1) * (bytesIn%1000)
        idx = idx % max_idx # Ou len(text) ?
        curr_chain = getChainFromTxt(text, idx)
        # Último dígito
        if d_1%2 != 0:
            curr_chain = '1' + curr_chain[1:]
        # Penúltimo dígito
        if d_2%2 != 0:
            curr_chain = reverseString(curr_chain)
        # Antepenúltimo dígito
        if d_3%2 != 0:
            curr_chain = invertBits(curr_chain)
        curr_chain = tabBinByte(curr_chain)
        # Troca aleatória de bits
        curr_chain = swapBits(curr_chain, d_1, d_2)
        if debugIdx: 
            print(bytesOut, end = '\t')
            print(f"d_1 = {d_1} i_1 = {i_1}", end=' ' + (i_1>=0)*' ')
            print(f'd_2 = {d_2} i_2 = {i_2}', end=' ' + (i_2>=0)*' ')
            print(f'd_3 = {d_3} i_3 = {i_3}')
        byteChain += curr_chain
        remaining_bytes -= 1
    return byteChain
        
```

# Sorteador de página de notícia

```python
main_url = "https://g1.globo.com/"
```

```python
def pickNewsPage():
    net_read = psutil.net_io_counters()
    bytesOut, bytesIn = net_read.bytes_sent, net_read.bytes_recv
    driver = Firefox()
    driver.get(main_url)
    links = driver.find_elements(By.CLASS_NAME, "feed-post-link")
    link_idx = (bytesOut%1000)%len(links)
    chosen_link = links[link_idx]
    href = chosen_link.get_attribute('href')
    driver.quit()
    return href
```

```python
pickNewsPage()
```

```python
def g1RandomNumberGenerator(numBytes):
    url = pickNewsPage()
    text = getTextFromG1(url)
    byteChain = getByteChain(text, numBytes)
    return byteChain
```

```python
def separateBytes(num):
    i = 0
    while i < len(num):
        print(num[i:i+8], end=" ")
        i += 8
    print("")
```

```python
randomNumber = g1RandomNumberGenerator(64)
separateBytes(randomNumber)
```

```python
def countOnesEachPos(number):
    count = [0] * 8
    for i in range(len(number)):
        if number[i] == '1':
            count[i%8] += 1
    return count
```
